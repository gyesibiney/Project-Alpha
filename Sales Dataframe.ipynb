```python
import pandas as pd
```

```python
data=[{'product_id':23, 'name':'computer', 'wholesale_price': 500, 'retail_price':1000, 'sales':100}, 
{'product_id':96, 'name':'Python Workout', 'wholesale_price': 35,'retail_price':75, 'sales':1000},
{'product_id':97, 'name':'Pandas Workout', 'wholesale_price': 35, 'retail_price':75, 'sales':500},
{'product_id':15, 'name':'banana', 'wholesale_price': 0.5,'retail_price':1, 'sales':200},
{'product_id':87, 'name':'sandwich', 'wholesale_price': 3,'retail_price':5, 'sales':300}]
```

```python
df=pd.DataFrame.from_records(data, index='product_id')
df
```

# Calculate the total revenue for each product

```python
#net_revenue_per_product = (retail_price - wholesale price) * sales
net_revenue_per_product=(df['retail_price']-df['wholesale_price'])*df['sales']
df['net_revenue_per_product']= net_revenue_per_product
df
```

# How much total revenue you received from all of these sales?

```python
#how much total revenue you received from all of these sales?
total_net_revenue= df['net_revenue_per_product'].sum()
print(f'total_net_revenue= {total_net_revenue}')
```

# What product is product retail price more than twice the wholesale price?

```python
# What product is product retail price more than twice the wholesale price?
product=df[df['retail_price']>df['wholesale_price']*2]
product

```

# How much did the store make from food vs. computers vs. books

```python

food= df[(df['name']=='banana')| (df['name']=='sandwich')]
food_rev =food['net_revenue_per_product'].sum()

computers= df[(df['name']=='computer')]
computer_rev =computers['net_revenue_per_product'].sum()

books= df[(df['name']=='Python Workout')| (df['name']=='Pandas Workout')]
books_rev =books['net_revenue_per_product'].sum()

print(f'food_revenue= {food_rev} vs computer_revenue= {computer_rev} vs books_revenue= {books_rev}')
```

## Because your store is doing so well, youâ€™re able to negotiate a 30% discount on the wholesale price of goods.


```python
#Calculate the new net revenue

discount_wholesale_price=df['wholesale_price']*0.7
df['discount_wholesale_price']=discount_wholesale_price

new_net_revenue_per_product=(df['retail_price']-df['discount_wholesale_price'])*df['sales']
df['new_net_revenue_per_product']= new_net_revenue_per_product
df
```

# taxes

```python
# taxes
df['revenue_@15%_sales_tax']= new_net_revenue_per_product*.85
df['revenue_@20%_sales_tax']= new_net_revenue_per_product*.80
df['revenue_@25%_sales_tax']= new_net_revenue_per_product*.75
df
```
